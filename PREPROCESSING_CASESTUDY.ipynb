##Handling Missing Values

```python
#Identify missing values in each column.
housing_data.isna().sum()
```

```python
#sale price is the target variable so we are removing the null rows
housing_data=housing_data.dropna(subset=['Sale Price'])
```

```python
from scipy.stats import skew
```

```python
skewness =housing_data['No of Bathrooms'].skew
skewness() #checking skew
```

```python
housing_data.loc[:,'No of Bathrooms'] = housing_data['No of Bathrooms'].fillna(housing_data['No of Bathrooms'].median()) #filling with median
```

```python
housing_data.loc[:,'Flat Area (in Sqft)'] = housing_data['Flat Area (in Sqft)'].fillna(housing_data['Area of the House from Basement (in Sqft)'] + housing_data['Basement Area (in Sqft)'])
```

```python
housing_data.loc[:,'Area of the House from Basement (in Sqft)']=housing_data['Area of the House from Basement (in Sqft)'].fillna(housing_data['Flat Area (in Sqft)']-housing_data['Lot Area (in Sqft)'])
```

```python
housing_data.isnull().sum()
```

```python
housing_data.loc[:,'Lot Area (in Sqft)']=housing_data['Lot Area (in Sqft)'].fillna(housing_data['Lot Area after Renovation (in Sqft)'])
```

```python
housing_data = housing_data.drop('No of Times Visited', axis=1)
```

```python
housing_data = housing_data.dropna(subset=['Zipcode'])
```

```python
housing_data = housing_data.drop(['Latitude', 'Longitude'],axis=1)
```

```python
housing_data = housing_data.drop(['Lot Area (in Sqft)','Renovated Year'],axis=1)
```

```python
housing_data = housing_data.dropna(subset=['Living Area after Renovation (in Sqft)'])
```

```python
housing_data=housing_data.drop(['ID','Area of the House from Basement (in Sqft)','Basement Area (in Sqft)'],axis=1)
```

```python
#handling outlier in no of bathrooom
skewness=housing_data['No of Bedrooms'].skew
skewness()
```

```python
q1 = np.quantile(housing_data['No of Bedrooms'],0.25)
q3 = np.quantile(housing_data['No of Bedrooms'],0.75)
iqr = q3 - q1
upper_bound = q3+(1.5*iqr)
lower_bound = q1-(1.5*iqr)

print(upper_bound)
print(lower_bound)
print(iqr)
```

```python
no_of_bedroom_outliers = []
for i in housing_data['No of Bedrooms']:
  if i > upper_bound or i < lower_bound:
    no_of_bedroom_outliers.append(i)
no_of_bedroom_outliers
```

```python
housing_data['No of Bedrooms']= housing_data['No of Bedrooms'].clip(lower = lower_bound,upper= upper_bound) #clipping data using IQR
```

```python
plt.boxplot(housing_data['No of Bedrooms'])
```

```python
no_of_bathroom_outliers = []
for i in housing_data['No of Bathrooms']:
  if i > upper_bound or i < lower_bound:
    no_of_bathroom_outliers.append(i)
no_of_bathroom_outliers
```

```python
#data is highly skewed so apply log transformation
#housing_data['Living Area after Renovation (in Sqft)']=np.log(housing_data['Living Area after Renovation (in Sqft)'])
```

```python
from scipy.stats.mstats import winsorize

# Calculate the winsorized values (capping at 5% from both ends)
housing_data['Living Area after Renovation (in Sqft)'] = winsorize(housing_data['Living Area after Renovation (in Sqft)'], limits=[0.05, 0.05])

# Check the skewness after winsorizing
skewness = housing_data['Living Area after Renovation (in Sqft)'].skew()
print(f"Skewness after winsorizing: {skewness}")

# Plot a boxplot to visualize the effect of winsorizing
plt.boxplot(housing_data['Living Area after Renovation (in Sqft)'])
plt.title('Living Area after Renovation (in Sqft) after Winsorizing')
plt.show()
```
