```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_auc_score
from sklearn.preprocessing import RobustScaler
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from xgboost import XGBClassifier
from sklearn.decomposition import PCA
```

```python
from sklearn.datasets import load_breast_cancer
data = load_breast_cancer()
```

```python
bc_data = pd.DataFrame(data.data, columns=data.feature_names)
bc_data['target'] = data.target
```

```python
for i in bc_data.columns:
  plt.figure(figsize=(10,5))
  sns.histplot(data=bc_data[i], kde=True)
  plt.title(f'Histogram of {i}')
  plt.show()
```

```python
y = bc_data.target
x = bc_data.drop('target', axis=1)

train_x, valid_x, train_y, valid_y = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('scaler', RobustScaler(), x.columns)
    ]
)
```

```python
log_reg_model = LogisticRegression(random_state=42)
dtc_model = DecisionTreeClassifier(random_state=42)
rfc_model = RandomForestClassifier(random_state=42)
xgbc_model = XGBClassifier(random_state=42)

models = [log_reg_model, dtc_model, rfc_model, xgbc_model]

```

```python
for model in models:
  pipeline = Pipeline(steps=[
  ('preprocessor', preprocessor),
  ('model', model)
  ])


  pipeline.fit(train_x, train_y)
  y_pred = pipeline.predict(valid_x)
  print(f'Model: {type(model).__name__}')
  print(f'Accuracy score: {accuracy_score(valid_y, y_pred)}')
```

```python
pca_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('pca', PCA())
])

```

```python
x_processed_pca = pca_pipeline.fit_transform(train_x)
pca = pca_pipeline.named_steps['pca']
plt.plot(np.cumsum(pca.explained_variance_ratio_), marker='o')
plt.xlabel('Number of components')
plt.ylabel('Cumulated explained variance')
plt.title('Components vs. Explained variance')
plt.grid(True)
plt.show()
```

```python
x_preprocessed = preprocessor.fit_transform(train_x)
pca = PCA(n_components=0.95, random_state=42)
x_pca = pca.fit_transform(x_preprocessed)
print('chosen components: ', pca.n_components_)
```

```python
for model in models:
    pipeline = Pipeline(steps=[
        ('preprocessor', preprocessor),
        ('pca', PCA(n_components=10, random_state=42)),
        ('model', model)
    ])
    pipeline.fit(train_x, train_y)
    y_pred = pipeline.predict(valid_x)
    print(f"Model: {type(model).__name__}")
    print(f"Accuracy: {accuracy_score(valid_y, y_pred)}")
```

### Model Performance Comparison

 a comparison of the accuracy scores for each model before and after applying PCA with 10 components:

| Model                | Accuracy (Before PCA) | Accuracy (After PCA) |
|----------------------|-----------------------|----------------------|
| Logistic Regression  | 0.982                 | 0.982                |
| Decision Tree        | 0.947                 | 0.947                |
| Random Forest        | 0.965                 | 0.947                |
| XGBoost              | 0.956                 | 0.965                |

In this case, applying PCA with 10 components did not significantly change the performance of most models, with Logistic Regression and Decision Tree having the same accuracy, while Random Forest accuracy decreased slightly and XGBoost accuracy increased slightly.
