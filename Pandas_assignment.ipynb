```python
import pandas as pd
import numpy as np

import seaborn as sns
```

1. Load the Titanic dataset and display the first 10 rows

```python
data=sns.load_dataset("titanic")
```

```python
data.head(10)
```

2. Find the number of missing values in each column

```python
data.isna().sum()
```

3. Find the percentage of passengers who survived, grouped by gender

```python
percentage=data.groupby("sex")["survived"].mean()*100
percentage.round(2)
```

4. Find the average, median, and standard deviation of passenger ages.


```python
age=data['age']
avr=age.mean()
med=age.median()
std_age=age.std()
```

```python
print(avr)
print(med)
print(std_age)
```

5. Count how many passengers were in each class (pclass).


```python
pclass=data['pclass']
pclass.value_counts()
```

6. Show all passengers who were female, under age 18, and survived.


```python
survived_female = data[(data['sex'] == 'female') & (data['age'] < 18) & (data['survived'] == 1)]
display(survived_female)
```

7. Replace missing age values with the median age

```python
age.fillna(med,inplace=True)
```

```python
age
```

8. Create a column age_group with values:
a. 'Child' for age < 18
b. 'Adult' for age 1860
c. 'Senior' for age > 60
d. Use: apply() or np.where() or custom function

```python
conditions = [data['age'] < 18, (data['age'] >= 18) & (data['age'] < 60), data['age']>60]
values = ['Child','Adult', 'Senior']
data['age_group'] = np.select(conditions, values, default='Unknown')
data.head(10)
```

9. Display the top 5 passengers who paid the highest fare.

```python
data = data.sort_values("fare",ascending=False)
data.head()
```

10.10. Find the average fare paid, grouped by both pclass and sex

```python
data.groupby(['pclass','sex'])["fare"].mean()
```
