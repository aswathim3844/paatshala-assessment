###Integrate Data:


Merge the datasets based on the country name

```python
data = pd.merge(country_population,literacy_rate, on='country', how='outer')


data = pd.merge(data, internet_users, on='country', how='outer')

data = pd.merge(data , country_gdp, on='country', how='outer')

display(data)
```

Load into SQLite:

Create an SQLite database


Load the final merged dataset into a database table


```python
import sqlite3

conn = sqlite3.connect("data")
cursor = conn.cursor()
```

```python
data.to_sql("final_data",conn,index=False)
```

```python
cursor.execute("""SELECT * FROM final_data LIMIT 5""")
rows = cursor.fetchall()

for row in rows:
    print(row)
```

a. Countries with highest internet penetration


```python
cursor.execute("""SELECT country, internet_penetration_rate
FROM final_data
ORDER BY internet_penetration_rate DESC
LIMIT 10""" )
rows = cursor.fetchall()
for i in rows:
  print(i)
```

b. Average literacy rate

```python
cursor.execute("select avg( literacy_rate)  from final_data")
print(cursor.fetchall())
```

c. Correlation between literacy rate and internet penetration

```python
cursor.execute("""SELECT
    (AVG(internet_penetration_rate * literacy_rate) - AVG(internet_penetration_rate) * AVG(literacy_rate)) /
    (SQRT((AVG(internet_penetration_rate * internet_penetration_rate) - AVG(internet_penetration_rate)*AVG(internet_penetration_rate)) *
          (AVG(literacy_rate * literacy_rate) - AVG(literacy_rate)*AVG(literacy_rate))))
    AS correlation
FROM final_data;""")

print(cursor.fetchall())
```

d. GDP per capita analysis

```python
cursor.execute("""
SELECT Country,
       ROUND(GDP / Population, 2) AS GDPPerCapita
FROM final_data

""")

print(cursor.fetchall())
```

e. Which countries have a GDP per capita above $10,000?

```python
cursor.execute("""
SELECT country from final_data where GDP  / population > 10000
    """
)
print(cursor.fetchall())
```

f. What is the total population covered in the dataset?

g. Which countries have the lowest literacy rates, and how does that impact internet
access?

```python
cursor.execute("""
SELECT country, literacy_rate, internet_penetration_rate
FROM final_data
ORDER BY literacy_rate ASC
""")

lowest_literacy_countries = cursor.fetchmany(10)


for row in lowest_literacy_countries:
    print(f"Country: {row[0]}, Literacy Rate: {row[1]}, Internet Penetration Rate: {row[2]}")
```

h. What are the top 5 wealthiest countries by total GDP, and how does that compare
with population size?


```python
cursor.execute("""
SELECT country , gdp , population , (gdp* population) as gdp_total from final_data
order by gdp_total desc""")
wealthiest_countries = cursor.fetchmany(5)


for row in wealthiest_countries :
    print(f"Country: {row[0]}, Literacy Rate: {row[1]}, Internet Penetration Rate: {row[2]}")
```

Find countries where internet users exceed 70% of the population.

```python
cursor.execute("""
SELECT country, internet_penetration_rate
FROM final_data
WHERE internet_penetration_rate > 70
ORDER BY internet_penetration_rate DESC
""")

countries_high_internet = cursor.fetchall()


for row in countries_high_internet:
    print(f"Country: {row[0]}, Internet Penetration Rate: {row[1]}")
```

j. What is the average GDP per capita for countries with internet penetration above
50%?

```python
cursor.execute("""
SELECT country, internet_penetration_rate, avg(gdp)
FROM final_data
WHERE internet_penetration_rate > 50
ORDER BY internet_penetration_rate DESC
""")
countries_penetration= cursor.fetchall()


for row in countries_penetration:
    print(f"Country: {row[0]}, Internet Penetration Rate: {row[1]},avg(gdp):{row[2]}")
```

How many countries have a literacy rate above 90%, and what is their average
internet penetration?

```python
cursor.execute("""
SELECT country, literacy_rate, avg(	internet_penetration_rate)
FROM final_data
WHERE literacy_rate > 90
ORDER BY literacy_rate DESC
""")
countries_penetration= cursor.fetchall()


for row in countries_penetration:
    print(f"Country: {row[0]}, Internet Penetration Rate: {row[1]},avg(internet_penetration_rate):{row[2]}")
```
